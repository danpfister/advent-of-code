```python
import numpy as np

inputfile = open("input.txt", "r")
inputdata = [line.strip() for line in inputfile.readlines()][0]
```

```python
inputdata = inputdata.split(",")
```

```python
invalids = 0
for pair in inputdata:
    start, end = pair.split("-")
    for i in range(int(start), int(end) + 1):
        current = str(i)
        # print(current)
        for j in range(1, len(current)):
            sub = current[0:j]
            found = True
            # print(f"sub {sub}, j {j}")
            if len(current)%j != 0: continue
            for k in range(0, len(current), j):
                if current[k: k+j] != sub:
                    found = False
            if found:
                print(f"match: {current}, {sub}")
                invalids += i
                break

print(invalids)
```
