```python
inputfile = open("input.txt", "r")
inputdata = [[int(c) for c in line.strip()] for line in inputfile.readlines()]
```

```python
part1 = sum([max(bank[:-1]) * 10 + max(bank[bank.index(max(bank[:-1]))+1:]) for bank in inputdata])
print(f"part 1: {part1}")
```

```python
sum = 0
for bank in inputdata:
    nr = ""
    ind = 0
    digits = 12
    for r in reversed(range(12)):
        curr = bank[ind:-r] if r > 0 else bank[ind:]
        m, i = max(curr), curr.index(max(curr))
        ind += i + 1
        # print(f"curr {curr}, m {m}, i {i}")
        nr += str(m)
    sum += int(nr)
print(f"part 2: {sum}")
```
