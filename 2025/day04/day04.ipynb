```python
import numpy as np
```

```python
inputfile = open("input.txt", "r")
inputdata = [[c for c in line.strip()] for line in inputfile.readlines()]
```

```python
data = np.asarray(inputdata)
```

```python
padded = np.pad(data, 1, 'constant', constant_values='.')
```

```python
h, w = padded.shape
cnt = 0

for i in range(h):
    for j in range(w):
        if padded[i, j] != '@': continue
        cnt += np.count_nonzero(padded[i-1:i+2, j-1:j+2] == '@') < 5
        
print(f"part 1: {cnt}")
```

```python
current = padded.copy()
next = current.copy()
cnt = 0

while True:
    changed = False
    for i in range(h):
        for j in range(w):
            if current[i, j] != '@': continue
            if np.count_nonzero(current[i-1:i+2, j-1:j+2] == '@') < 5:
                next[i, j] = '.'
                changed = True
    current = next
    if not changed: break

print(f"part 2: {np.count_nonzero(current != padded)}")
```
