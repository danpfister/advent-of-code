```python
inputfile = open("input.txt", "r")
inputdata = [line.strip() for line in inputfile.readlines()]
```

```python
ranges = []
ings = []

for line in inputdata:
    if "-" in line:
        ranges.append([int(c) for c in line.split('-')[0:2]])
    elif line != "":
        ings.append(int(line))
```

```python
fresh = 0

for i in ings:
    for a, b in ranges:
        if i >= a and i <= b:
            fresh += 1
            break

print(f"part 1: {fresh}")
```

```python
sorted_ranges = sorted(ranges, key=lambda x: x[0])
```

```python
merged = [sorted_ranges[0]]
for a, b in sorted_ranges[1:]:
    c, d = merged[-1]
    if a <= d:
        merged[-1] = [c, max(b, d)]
    else:
        merged.append([a, b])

print(f"part2: {sum([b - a + 1 for a, b in merged])}")
```
