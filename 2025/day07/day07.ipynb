```python
import numpy as np

inputfile = open("input.txt", "r")
inputdata = [[c for c in line.strip()] for line in inputfile.readlines()]
```

```python
data = np.asarray(inputdata)
```

```python
s = np.argwhere(data == "S")[0]
data[s[0], s[1]] = "|"
```

```python
splits = 0
dag = {}
for layer in range(data.shape[0]-1):
    for i in range(data.shape[1]):
        if data[layer, i] == "|":
            if data[layer+1, i] == "^":
                data[layer+1, i-1] = "|"
                data[layer+1, i+1] = "|"
                splits += 1
                dag.update({(layer, i): [(layer+1, i-1), (layer+1, i+1)]})
            else:
                data[layer+1, i] = "|"
                dag.update({(layer, i): [(layer+1, i)]})
```

```python
print(f"part 1: {splits}")
```

- view the places where the beam has been as nodes in a DAG, with the next beam locations as successors of the nodes
- store in a dictionary
- it is already topologically sorted

```python
paths = {}
paths[(s[0], s[1])] = 1
```

```python
# give source a value of 1, to each successor node, add the value of the current node
for curr, succs in dag.items():
    for succ in succs:
        paths[succ] = paths.get(succ, 0) + paths[curr]
```

```python
timelines = sum([v for k, v in paths.items() if k[0] == data.shape[0]-1])
print(f"part 2: {timelines}")
```
